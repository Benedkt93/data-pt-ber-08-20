#### Using df.apply

```python
# Apply Function
# Used when a function is not supposed to run at a whole dataframe at once but row by row or column by column or element by element
# Using df.apply
# Apply works on Series not on Frame (?!)
# The equivalent for apply for lists is map,reduce, filter & apply function is also structured the same way
# As well as map,reduce, filter also apply could be done by using a for-loop
# Structure: df.apply(function,axis)
```

```python
# as upper is tied to a class --> Method, no function
df['Manufacturer'].str.upper()
# or
df["Manufactures"].apply(str.upper)
# alternative
# we hae x as it is a string already
df["Manufactures"].apply(lambda x: x.upper())

# Apply works with Lambda functions
```

```python
# Replace all  "a" by "b"

df["Manufactures"].apply(lambda x: x.replace("a", "b"))
```

```python
# in order to apply the apply to a dataframe the alternative way is to define a function
def replace(df):
    return df.str.replace("e","b")

# after define this function you can run

df[["Manufacturer","Fuel Type"]].apply(replace)
```

```python
subset = df[['Manufacturer', 'Vehicle Class']]
for i in range(len(subset.columns)):
    print(df[subset.columns[i]].apply(lambda x: x.replace('e', 'b')))
```

```python
# Investigating and replacing NUll values
df. isnull()
# or
pd.isnull(df)
# second one if applying to a single cell is common
# when applying to whole dataframe, than use first one
# same structure stays and it replaces all value by chekcing if it is empty or not and will replace it by True or False
```

```python
# find amount of null values of each column, sums up the True values for missing values
df.isnull().sum()

# percentage
df.isnull().sum() / len(df)
# wieso nicht count?! Da count nur die die nicht missing values anzeigt
# Trotzdem vorsichtig sein mit len in data frame bc it usually destroys dataframe and just shows one value
```

```python
# Subsetting:  Which columns do not have missing values 
null_cols = df.isnull().sum() == 0
# Which colums do have missing values
null_cols = df.isnull().sum() > 0

# but up to know still all colmuns are shown, therefore we just want to have the ones with missing values, by where clause
null_cols = df.isnull().sum()
null_cols(null_cols > 0) #---> bracket represents where clause 
# now just all variables with missing values and their amount
```

```python
# Now all null cols with no missing values
null_cols = df.isnull().sum()
null_cols(null_cols == 0)
```

```python
# List of columns with more than 10k missing values
null_cols = df.isnull().sum()
null_cols(null_cols > 10000)


```

```python
# Dropping column evMotors, mfrCode
df.drop(["evMotors", "mfrCode"], axis =1)
```

```python
# Drop all columns which have more than 10k missing values
# (wenn sum usw, wir haben kein Datframe mehr aber eine Serie) Wir haben eine Serie aus Index und einer Serie
cols_missing_10000 = list(null_cols(null_cols > 10000). index)
print (cols_missing_1000)
```

```python
df = df.drop(cols_missing_1000, axis = 1)
# hier originally is overridden
```
