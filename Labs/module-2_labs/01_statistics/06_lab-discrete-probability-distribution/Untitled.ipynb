### Bernoulli

```python
p = 0.5
bern= scipy.stats.bernoulli(p)
```

```python
bern.pmf(0)
```

```python
bern.pmf(1)
```

```python
bern.pmf(4)
```

```python
bern.cdf(0)
```

```python
bern.cdf(1)
```

```python
bern.cdf(2)
```

```python
numrange = range(10)
plt.plot(numrange, bern.pmf(numrange), "o")
```

```python
plt.plot(numrange, bern.cdf(numrange))
```

```python
import matplotlib.pyplot as plt
X = bern.rvs(size = 100)
plt.hist(X)
```

### Binominal

```python
from scipy.stats import binom
p= 0.5
n = 2
binominal = scipy.stats.binom(n,p)
```

```python
binominal.pmf(0)
```

```python
plt.plot(numrange, binominal.pmf(numrange), "o")
```

```python
binominal.cdf(1)
```

```python
plt.plot(numrange, binominal.cdf(numrange))
```

```python
X = binominal.rvs(size = 100)
plt.hist(X)
```

```python
p= 0.5
n = 20
binominal = scipy.stats.binom(n,p)
X = binominal.rvs(size = 100)
plt.hist(X)
# 100 times we make 20 tosses and X are the heads of the 20 times
```

```python
numrange = range (-10, 30)
plt.vlines(numrange, 0, binominal.pmf(numrange))
```

```python
# Bonominal is that mass goes to the middle over time
# Uniform was flat
# Bernouli was seperated to the two values
# numrange just shows list of numbers and I basically just want to see how likely it is that a number in range is the ampunt of heads i get
# Theoretical prediction
```
