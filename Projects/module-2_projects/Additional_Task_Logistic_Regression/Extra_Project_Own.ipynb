```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
logs_web = pd.read_csv('logs_web.csv')
logs_web.head()
```

```python
logs_web.shape
```

```python
logs_notifications = pd.read_csv('logs_notifications.csv')
logs_notifications.head()
```

```python
logs_notifications.shape
```

```python
df_new = logs_notifications.merge(logs_web, on='ip_anonymized', how='left')
df_new
```

```python
df_new['YesNo'] = np.where(df_new['message_y'].isnull(),0,1)
df_new
```

```python
df_dum1 = pd.get_dummies(df_new['message_x'],drop_first = True)
df_dum2 = pd.get_dummies(df_new['language_x'],drop_first = True)
df_dum3= pd.get_dummies(df_new['browser_x'],drop_first = True)
```

```python
df_final = df_new.merge(df_dum1, left_index = True, right_index = True)
df_final
```

```python
df_final2=df_final.merge(df_dum2, left_index = True, right_index = True)
df_final2
```

```python
df_final3=df_final2.merge(df_dum3, left_index = True, right_index = True)
df_final3
```

```python
Y = df_final3["YesNo"]
X = df_final3[[1,2]]
X = sm.add_constant(X)
model = sm.Logit(Y, X)
results = model.fit()
results.summary()
```

```python
Y = df_final3["YesNo"]
X = df_final3[["de","en","es","fr","ru"]]
X = sm.add_constant(X)
model = sm.Logit(Y, X)
results = model.fit()
results.summary()
```

```python
Y = df_final3["YesNo"]
X = df_final3["firefox"]
X = sm.add_constant(X)
model = sm.Logit(Y, X)
results = model.fit()
results.summary()
```
